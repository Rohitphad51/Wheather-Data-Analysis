```python
import pandas as pd

# Load the dataset
df = pd.read_csv("weather_dataset.csv")  #this fun read data from csv file 

# Display the first few rows of the dataset
print(df.head()) # this prints the first 5 rows of the daatframe

# Get basic information about the data (data types, missing values, etc.)
print(df.info()) #print the details rows columns names data types number of non null vlaues 

# Descriptive statistics of numerical columns
print(df.describe()) # count average standard deviation min max values 
```

```python
# Handle missing data (you can drop or fill)
df = df.dropna()  # Drop rows with missing values
# or 
# df = df.fillna(method='ffill')  # You can use forward fill to handle missing values

print(df.isnull().sum())  # Check if any missing values remain
```

```python
df['Date'] = pd.to_datetime(df['Date']) # this convert the date column in the dataframe into a proper datetime format using

# Check the updated data types
print(df.dtypes) # this line prints the data type of each columns in the dataframe
```

```python
    import matplotlib.pyplot as plt

# Plot temperature over time
plt.figure(figsize=(12, 6)) # this create a figsize 
plt.plot(df['Date'], df['Temperature_C'], color='orange')  # this create a line plot using the date column on x axix an temp y axis 
plt.title('Temperature Over Time') # this set the title 
plt.xlabel('Date') 
plt.ylabel('Temperature_C (°C)')
plt.grid(True)
plt.show()
```

```python
# Plot humidity over time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Humidity_%'], color='blue')  # x  axis date y axis humidity
plt.title('Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Humidity_% (%)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Scatter plot for Temperature vs Humidity
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temperature_C', y='Humidity_%', data=df) # x axis temp and  y aixs humidity 
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature_C (°C)')
plt.ylabel('Humidity_% (%)')
plt.show()
```

```python
# Weather condition count
weather_counts = df['Weather_Condition'].value_counts() # this counts how many times each weather condition appears in the weather condition 
print(weather_counts) # this prints result

# Bar plot of weather conditions
plt.figure(figsize=(10, 5))
sns.countplot(x='Weather_Condition', data=df, order=weather_counts.index) # this uses seaborn countplot to create a barchat diff  wc x y show weather types 
# data=df is the dataframe using ,
plt.title('Weather Condition Frequency')
plt.xticks(rotation=45) # this rotates a x axis labels by 45 degree to make them more readble
plt.tight_layout() # this automatically adjust the sapacing the plot
plt.show()
```
